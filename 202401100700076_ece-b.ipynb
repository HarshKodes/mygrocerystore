Grocery Store Calculator


```python
from tabulate import tabulate

print ("Welcome to Harsh's store.")
print ("This is a program for calculating your total bill.")
choice=1

name = []
price = []
quantity = []
gst = []
total_cost = []

while choice != 0:
    if choice == 1:
        name1 = input("Enter the name of the item: ")  # taking inputs
        price1 = float(input("Enter the price of the item: "))
        quantity1 = int(input("Enter the quantity of the item: "))

        gst1 = 0.10 * (price1 * quantity1)
        total_cost1 = (price1 * quantity1) + gst1

        name.append(name1)  # adding items
        price.append(price1)
        quantity.append(quantity1)
        gst.append(gst1)
        total_cost.append(total_cost1)

    else:
        print("Invalid choice. Please try again.")

    print("\nEnter 1 for adding more items.")
    print("Enter 0 for calculating final cost.")
    choice = int(input("Enter your choice: "))
    print("\n")

# printing final bill in tabular form
data = []
for i in range(len(name)):
    data.append([name[i], price[i], quantity[i], round(gst[i], 2), round(total_cost[i], 2)])

print("\nFinal Bill:")
print(tabulate(data, headers=["Item", "Price", "Quantity", "GST", "Total Cost"], tablefmt="grid"))
print("\nTotal Bill Amount: ₹", round(sum(total_cost), 2))

print ("\nCongratulations on winning a 10% discount as a part of our Mahakumbh sale.")
print ("Money saved : ₹", round(sum(total_cost)*0.1, 2))
print ("Total amount saved : ₹", round(sum(total_cost)*0.9, 2))

```
