```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the .xlsx file
try:
    data = pd.read_excel("Case_Study_5_B.xlsx")  # Replace with your actual file
except FileNotFoundError:
    print("Error: File not found. Please make sure the file is in the correct location and the name is accurate.")
    exit()

# Validate 'Marks' column exists
if 'Marks' not in data.columns:
    print("Error: 'Marks' column not found in the Excel file.")
    exit()

# Calculate percentage (out of 15 marks)
data['Percentage'] = (data['Marks'] / 15) * 100

# Categorize students
high_achievers = data[data['Percentage'] > 75]
average_students = data[(data['Percentage'] >= 60) & (data['Percentage'] <= 75)]
low_achievers = data[data['Percentage'] < 60]

# Print categorized data
print("Students with greater than 75% marks:")
print(high_achievers)
print("\nStudents with marks between 60% and 75%:")
print(average_students)
print("\nStudents with less than 60% marks:")
print(low_achievers)

# Histogram plot
plt.figure(figsize=(10, 6))
plt.hist(data['Percentage'], bins=range(0, 101, 5), edgecolor='black', color='blue')
plt.xlabel("Percentage")
plt.ylabel("Number of Students")
plt.title("Histogram Plot")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()

# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(high_achievers.index, high_achievers['Percentage'], label='>75%', color='blue')
plt.scatter(average_students.index, average_students['Percentage'], label='6075%', color='green')
plt.scatter(low_achievers.index, low_achievers['Percentage'], label='<60%', color='red')
plt.xlabel("Students")
plt.ylabel("Percentage")
plt.title("Scatter Plot")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
